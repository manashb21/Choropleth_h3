```python
import geopandas as gpd
import numpy as np
import pandas as pd
import h3
```

```python
fire_ignitions = gpd.read_file(r'Fire_Ignition_Locations\Fire_Ignition_Locations.shp')
```

```python
numeric_columns = (fire_ignitions.select_dtypes(include = np.number).columns.to_list())
fire_ignitions[numeric_columns].head()
```

```python
fire_ignitions.dropna(inplace = True)
fire_ignitions = (fire_ignitions[['FIRE_ID','LATITUDE_D','LONGITUDE_']])
fire_ignitions.head()
```

```python
h3_res = 5

def geo_to_h3(row):
    return h3.geo_to_h3(lat = row.LATITUDE_D, lng = row.LONGITUDE_, resolution = h3_res)
```

```python
fire_ignitions['h3_cell'] = fire_ignitions.apply(geo_to_h3, axis =1 )
fire_ignitions.head()
```

```python
fire_ignitions = (fire_ignitions[['FIRE_ID','h3_cell']])
fire_ignitions.rename(columns = {'FIRE_ID':'ignition_id'}, inplace = True)
fire_ignitions.head()
```

```python
#grouping done on basis of h3_cell, count all fire ignitions points located in h3 cell 
fire_ignitions_g = (fire_ignitions
                    .groupby('h3_cell')
                    .ignition_id
                    .agg(list)
                    .to_frame('ids')
                    .reset_index())
fire_ignitions_g['count'] = (fire_ignitions_g['ids'].apply(lambda ignition_ids:len(ignition_ids)))
fire_ignitions_g.head()
#fire_ignitions_g.sort_values('count', ascending = False)
```
